```python
import numpy as np
features=np.array([[0,0],[0,1],[1,0],[1,1]])
labels=np.array([0,0,0,1])
epoch=int(input("enter epocj"))
threshold=float(input('enter threshhold:'))
learning_rate=float(input('enter learning_rate:'))
a=[]
a.append(float(input('enter weith for x0:')))
a.append(float(input('enter weith for x1:')))
w=np.array(a)
for j in range(0, epoch):
    print("Epoch:", j)
    global_delta = 0

    for i in range(0, features.shape[0]):
        # Extract actual label and feature instance
        actual = labels[i]
        instance = features[i]
        x0 = instance[0]
        x1 = instance[1]

        print("X0:", x0)
        print("X1:", x1)

        # Compute the sum unit
        sum_unit = x0 * w[0] + x1 * w[1]
        print("Sum Value:", sum_unit)

        # Determine the output based on the threshold
        fire = 1 if sum_unit > threshold else 0

        # Compute the delta (error)
        delta = actual - fire
        print("Delta:", delta)

        # Accumulate the absolute delta to track global error
        global_delta += abs(delta)
        print("Prediction:", fire, "whereas actual was", actual, "(Error:", delta, ")")

        # Update weights
        w[0] = w[0] + delta * learning_rate * x0
        w[1] = w[1] + delta * learning_rate * x1

        print("--------------------")

    # If there is no error in the current epoch, terminate training
    if global_delta == 0:
        break
```
